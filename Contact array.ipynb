```python
import numpy as np
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt

x = 3.27
age_groups_x = ["0-9", "10-18", "19-64", "65+"]
age_groups_y = ["65+", "19-64", "10-18", "0-9"]

data = np.array([[0.5, 0.8, 1.05, 1.85],[5.63, 5.7, 7.93, 3.87], [1.29, 8.0, 1.24, 0.64],[5.1, 1.16, 1.1, 0.36]])
ax = plt.axes()
heat_map = sb.heatmap(data, annot=True, cbar=False, cmap = "BuGn")
ax.set_xticks(np.arange(len(age_groups_x)))
ax.set_yticks(np.arange(len(age_groups_y)))
ax.set_xticklabels(age_groups_x)
ax.set_yticklabels(age_groups_y)
plt.xlabel("Age group of Susceptible")
plt.ylabel("Age group of Infected Contact")
ax.set_title("Contact Array Between age groups")
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax.get_yticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.show()


#R_0 = beta / recovery rate (y)

y=0.26
data = np.array([[0.5/y, 0.8/y,1.05/y,1.85/y],[5.63/y,5.7/y,7.93/y,3.87/y], [1.29/y,8.0/y,1.24/y,0.64/y],[5.1/y,1.16/y,1.1/y,0.36/y]])
ax = plt.axes()
heat_map = sb.heatmap(data, annot=True, cbar=False, cmap = "BuGn")
plt.xlabel("Age group of Susceptible")
plt.ylabel("Age group of Infected Contact")
ax.set_xticks(np.arange(len(age_groups_x)))
ax.set_yticks(np.arange(len(age_groups_y)))
ax.set_xticklabels(age_groups_x)
ax.set_yticklabels(age_groups_y)
plt.xlabel("Age group of Susceptible")
plt.ylabel("Age group of Infected Contact")
ax.set_title("Placeholder Beta Array Between age groups")
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax.get_yticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.show()
```

```python
#Therefore the scaling factor is 0.0499
x = 0.0499
y=0.26
age_groups_x = ["0-9", "10-18", "19-64", "65+"]
age_groups_y = ["65+", "19-64", "10-18", "0-9"]
data = np.array([[x*0.5/y, x*0.8/y,x*1.05/y,x*1.85/y],[x*5.63/y,x*5.7/y,x*7.93/y,x*3.87/y], [x*1.29/y,x*8.0/y,x*1.24/y,x*0.64/y],[x*5.1/y,x*1.16/y,x*1.1/y,x*0.36/y]])
ax = plt.axes()
heat_map = sb.heatmap(data, annot=True, cbar=False, cmap = "BuGn", yticklabels=(4,3,2,1), xticklabels=(1,2,3,4))
plt.xlabel("Age group of Susceptible")
plt.ylabel("Age group of Infected Contact")
ax.set_title("Beta Array Between age groups")
ax.set_xticks(np.arange(len(age_groups_x)))
ax.set_yticks(np.arange(len(age_groups_y)))
ax.set_xticklabels(age_groups_x)
ax.set_yticklabels(age_groups_y)
plt.xlabel("Age group of Susceptible")
plt.ylabel("Age group of Infected Contact")
ax.set_title("Beta Array Between age groups")
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax.get_yticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.show()
```
